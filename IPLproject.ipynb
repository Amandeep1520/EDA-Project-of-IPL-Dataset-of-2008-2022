```python
#loading the required Libraries
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading IPL match dataset (2008-2022)
ipl = pd.read_csv("C:/Users/amandeep/Downloads/IPL_Matches_2008_2022.csv")
```

```python
# First 5 records of the dataset
ipl.head()
```

```python
# Number of rows and columns in dataset 
ipl.shape
```

```python
# Getting the frequency of most man of the match awards.
ipl['Player_of_Match'].value_counts()
```

```python
# Getting Top 10 player with the maximum number of man of the match awards.
ipl['Player_of_Match'].value_counts()[0:10]
```

```python
#Getting Top 5 player with the maximum number of man of the match awards.
ipl['Player_of_Match'].value_counts()[0:5]
```

```python
list(ipl['Player_of_Match'].value_counts()[0:5].keys())
```

```python
plt.bar(list(ipl['Player_of_Match'].value_counts()[0:5].keys()),list(ipl['Player_of_Match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
# Finding out the number of toss win by each team
ipl['TossWinner'].value_counts()
```

```python
# Getting the records where a team won batting first
batting_first = ipl[(ipl['WonBy'] != 'Wickets')]
```

```python
batting_first.head()
```

```python
plt.hist(batting_first['Margin'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
# Finding number of wins eact team after batting first.
batting_first['WinningTeam'].value_counts()
```

```python
# Making a bar plot for the top 3 teams with most wins after batting first.
plt.figure(figsize=(8,6))
plt.bar(list(batting_first['WinningTeam'].value_counts()[0:3].keys()),list(batting_first['WinningTeam'].value_counts()[0:3]),color=['Blue','Yellow',
                                                                                                                                   'Red'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['WinningTeam'].value_counts()), labels=list(batting_first['WinningTeam'].value_counts().index), autopct='%0.1f%%')
plt.show()
```

```python
# Getting the records where a team won bowling first
bowling_first = ipl[(ipl['WonBy'] != 'Runs')]
```

```python
bowling_first.head()
```

```python
# Finding number of wins eact team after bowling first.
bowling_first['WinningTeam'].value_counts()
```

```python
# Making a bar plot for the top 3 teams with most wins after bowling first.
plt.figure(figsize=(8,6))
plt.bar(list(bowling_first['WinningTeam'].value_counts()[0:3].keys()),list(bowling_first['WinningTeam'].value_counts()[0:3]),color=['Purple','Blue',
                                                                                                                                   'Red'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(list(bowling_first['WinningTeam'].value_counts()), labels=list(bowling_first['WinningTeam'].value_counts().index), autopct='%0.1f%%')
plt.show()
```
